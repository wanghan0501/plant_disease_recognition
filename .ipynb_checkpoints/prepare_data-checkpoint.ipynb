```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
import shutil 

%matplotlib inline
```

```python
path = 'data/ai_challenger_pdr2018_trainingset_20180905/AgriculturalDisease_trainingset/AgriculturalDisease_train_annotations.json'
disease_data = pd.read_json(path)
image_prefix = 'data/ai_challenger_pdr2018_trainingset_20180905/AgriculturalDisease_trainingset/images'
```

```python
for idx, row in disease_data.iterrows():
    if row.disease_class == 5:
        print(row.image_id)
        shutil.copyfile(os.path.join(image_prefix, row.image_id), os.path.join(
            '../Self-Attention-GAN/data/apple/cedar_apple_rust_serious/', row.image_id))
        plt.imshow(Image.open(os.path.join(image_prefix, row.image_id)))
        plt.show()
```
